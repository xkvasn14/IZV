# Hypotéza 1
### Na silnicích první třídy se při nehodách umíralo se stejnou pravděpodobností jako dálnicích.
načteme knihovny pandas a scipy

```python
import pandas as pd
import scipy.stats as st
```

Načteme Dataframe z accidents.pkl.gz

```python
df = pd.read_pickle("accidents.pkl.gz")
df
```

Vytvoříme nové sloupce pro úmrtí na silnici, pro silnici první třídy a dálnici.
Vyloučíme možnost neúmrtí na silnici.
uložíme nehody na silnicích první třídy
uložíme nehody na dálnicích

```python
df['umrti'] = df['p13a'] > 0
df['prvnitrida'] = df['p36'] == 1
df['dalnice'] = df['p36'] == 0
```

Vytvoříme si kontingenční tabulky pomocí funkce crosstab z knihovny pandas.

```python
ct = pd.crosstab(df['umrti'], df['prvnitrida'])
dt = pd.crosstab(df['umrti'], df['dalnice'])
ct
```

```python
dt
```

Pro tyto tabulky vypočítáme chi test pomocí funkce chi2_contingency z knihovny scipy. Tato funkce spočítá  statistiku a p-hodnotu pro test nezávislosti v kontingenční tabulce. Tato funkce vrací více hodnot, nás zajímá hodnoty expected array

```python
prvnitrida = st.chi2_contingency(ct)
prvnitrida
```

```python
dalnice = st.chi2_contingency(dt)
dalnice
```

```python
st.stats.ttest_ind(dalnice[3],prvnitrida[3])
```

pvalue nám ukazuje, že hodnoty nejsou v vybraném prahu spolehlivosti 5%

# Hypotéza 2
### Při nehodách vozidel značky Škoda je škoda na vozidle nižší než při nehodách vozidel Audi.
načteme Dataframe

```python
df = pd.read_pickle("accidents.pkl.gz")
df
```

upravená cena škod z stokorun na koruny (není zapotřebí, pouze pro pohodlí uživatele)

```python
df['p53'] = df['p53'].astype(int).mul(100)
```

Vytvoříme nové sloupce pro cenu nehod, a jestli účastněné značky aut jsou audi nebo škoda

```python
df['cena'] = df['p53']
df['audi'] = df['p45a'] == 2
df['skoda'] = df['p45a'] == 39
```

```python
skoda = pd.crosstab(df['cena'], df['skoda'])
audi = pd.crosstab(df['cena'], df['audi'])
skoda
```

```python
audi
```

použijeme test ttest_ind z knihovny scipy. Podle hodnoty statistic zjistíme jak moc se liší škody nehod mezi značkami škoda a audi

```python
st.stats.ttest_ind(skoda[1],audi[1])
```

## Závěr hypotézy 2
hodnota statistic=3.35904641, z toho plyne že škody nehod na audi byly až 3.5x vyšší než na Škodách.
